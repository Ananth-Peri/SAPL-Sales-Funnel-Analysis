```python
# Reading the respective excel 'csv' files

city=pd.read_csv("city_table.csv")
hostel=pd.read_csv("hostel_table.csv")
lead=pd.read_csv("lead_visit_table.csv")
status=pd.read_csv("status_table.csv")
```

```python
# Renaming the column to match the name in other tables

# Here, the 'id' is a unique row identifier, hence changing the name to 'row_id'; not to be confused with 'id' in hostel table

lead.rename(columns={'sustatus_id':'substatus_id', 'id':'row_id', 'cityid':'city_id'}, inplace=True);lead
```

```python
# I'm merging the 'lead' with 'city' using left join on 'city_id', i.e. all the data points in 'lead' will be returned, along with matched 'city' values 

lead_city=pd.merge(lead, city, on="city_id", how="left"); lead_city
```

```python
# Joining the above merged table with 'status' on 'substatus_id' using inner join, since there are no missing values in the table

lead_city_status=pd.merge(lead_city, status, on="substatus_id", how="inner");lead_city_status
```

```python
# lead_city_status[lead_city_status["lead_id"]==1001]
```

```python
# Dropping the columns not required for analysis

lead_city_status.drop(lead_city_status.columns[[0,1,3,4]], axis=1, inplace=True); lead_city_status
```

```python
# Creating a new table from by using iloc and extracting the required columns

table_1=lead_city_status.iloc[:,[2,3,4]]; table_1
```

```python
# Creating a pivot table from the above dataframe

pivot_table_1=pd.pivot_table(table_1, columns="substatus_name", index="city", aggfunc= np.size, fill_value=0); pivot_table_1
```

```python
# Creating a new dataframe for the pivot with revised columns index

pivot_table=pivot_table_1.iloc[:,[24,21,15,20,18,6,16,23,19,7,13,14,9,10,11,13,12,0,4,1,2,3,5,22,17,25]]; pivot_table
```

```python
# Renaming the column axis of the pivot table

pivot_table.set_axis(["signup", "mob_upd", "ema_upd", "id_upd", "host_pref", "poli_verf", "ema_ver"
                    , "otp_ver", "id_appr", "verf_lead", "call_pend", "connect", "visit_sch",
                    "visit_canc", "visit_done", "visit_resch", "visit_miss", "agg_init", "nego_stg", "agg_shar",
                    "agg_sign", "agg_edit", "pay_rcvd", "onb_init", "host_pref", "stud_shif"], axis="columns", inplace=True)
```

```python
# Final pivot table with all the stages of the funnel

pivot_table
```
